# Missing Data

Let's show a few convenient methods to deal with Missing Data in pandas:

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'A':[1,2,np.nan],
                  'B':[5,np.nan,np.nan],
                  'C':[1,2,3]})
```

```python
df
```

```python
df.dropna() ## By default it check for rows
```

```python
df.dropna(axis=1) ## Here, its gonna check for columns
```

```python
df
```

```python
df.dropna(thresh=2)  ## keep only those rows with atleast 2 nan values which means the cuttoff for checking nan values is 2
```

```python
df.dropna(thresh=3)
```

```python
df.dropna(thresh=1)
```

```python
df.fillna(value='FILL VALUE')
```

```python
df.fillna(value='0', inplace=True)
```

```python
df
```

```python
df['A'].fillna(value=df['A'].mean())
```

```python
df["C"].mean()
```
