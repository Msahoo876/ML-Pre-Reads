```python
Grading system

marks > 90 and <100 (A grade)
marks >80 and <90    B grade
marks>70 and <80     B+ grade
marks >60 and <70    C grade
marks>50 and <60 D grade
```

```python
mark = int(input("Enter your marks to check your grade"))
if (mark>90 and mark<100):
	print("A")
elif (mark>80 and mark<90):
	print("B")
elif (mark>70 and mark<80):
	print("B+")
elif (mark>60 and mark<70):
	print("C")
elif (mark>50 and mark<60):
	print("D")
else:
  print("Invalid marks")
```
