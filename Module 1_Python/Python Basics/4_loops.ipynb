```python
for i in range(1,22): 
  if i%2==0: 
    print(i ,"= Fizz, only div by 2") 
  elif i%3==0:
    print(i ,"= Dizz, only div by 3") 
  elif i%2==0 and i%3==0: 
    print(i, "= Fizzdizz, div by 2 and 3 both")
```
